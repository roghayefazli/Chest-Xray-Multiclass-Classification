```python
import os
import shutil
import numpy as np

source_dir = "Lung X-Ray Image"
train_dir = "Lung X-Ray Imagedata_split/train"
test_dir = "Lung X-Ray Imagedata_split/test"

classes = ["Normal", "Lung_Opacity", "Viral Pneumonia"]

# ایجاد فولدرهای جدید
for c in classes:
    os.makedirs(os.path.join(train_dir, c), exist_ok=True)
    os.makedirs(os.path.join(test_dir, c), exist_ok=True)

# تقسیم داده‌ها
split_ratio = 0.8  # 80% train 20% test

for c in classes:
    class_path = os.path.join(source_dir, c)
    files = os.listdir(class_path)
    np.random.shuffle(files)

    split_index = int(len(files) * split_ratio)
    train_files = files[:split_index]
    test_files = files[split_index:]

    # انتقال فایل‌ها
    for f in train_files:
        shutil.copy(os.path.join(class_path, f), os.path.join(train_dir, c))

    for f in test_files:
        shutil.copy(os.path.join(class_path, f), os.path.join(test_dir, c))

print("Done! Dataset split into train/test.")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import os

# مسیر داده‌ها
train_dir = 'Lung X-Ray Imagedata_split/train'
val_dir = 'Lung X-Ray Imagedata_split/test'

# ImageDataGenerator با augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

# Loader داده‌ها
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# ساخت مدل ResNet50
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_model.trainable = False  # freeze base model initially

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
output = Dense(3, activation='softmax')(x)  # سه کلاس

model = Model(inputs=base_model.input, outputs=output)

# Compile
model.compile(optimizer=Adam(learning_rate=1e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
checkpoint = ModelCheckpoint('best_model.h5', monitor='val_accuracy', save_best_only=True, verbose=1)

# آموزش اولیه
history = model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr, checkpoint]
)

# Fine-tuning: باز کردن بعضی لایه‌ها
base_model.trainable = True
for layer in base_model.layers[:-50]:  # freeze پایین‌ترین لایه‌ها
    layer.trainable = False

model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history_finetune = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr, checkpoint]
)

# ارزیابی مدل
val_generator.reset()
Y_pred = model.predict(val_generator)
y_pred = np.argmax(Y_pred, axis=1)
y_true = val_generator.classes

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=val_generator.class_indices.keys(),
            yticklabels=val_generator.class_indices.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Classification report
print(classification_report(y_true, y_pred, target_names=val_generator.class_indices.keys()))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
import os

# === 1. بارگذاری مدل ===
model = load_model('best_model.h5')
print("✅ مدل بارگذاری شد.")

# === 2. تنظیمات داده ===
val_dir = 'Lung X-Ray Imagedata_split/test'  
IMG_SIZE = (224, 224)
BATCH_SIZE = 1  # برای دسترسی آسان به تصاویر فردی

# بدون shuffle و با batch_size=1 برای کنترل دقیق
val_datagen = ImageDataGenerator(rescale=1./255)
val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

class_names = list(val_generator.class_indices.keys())
total_samples = val_generator.samples

# === 3. پیش‌بینی روی تمام تصاویر ===
print("در حال پیش‌بینی روی تمام تصاویر...")
val_generator.reset()
all_preds = []
all_true = []

for i in range(total_samples):
    img, label = next(val_generator)
    pred = model.predict(img, verbose=0)
    all_preds.append(np.argmax(pred))
    all_true.append(np.argmax(label))

all_preds = np.array(all_preds)
all_true = np.array(all_true)

# === 4. پیدا کردن اندیس‌های درست و غلط ===
correct_indices = np.where(all_preds == all_true)[0]
wrong_indices = np.where(all_preds != all_true)[0]

# print(f"✅ تعداد پیش‌بینی‌های درست: {len(correct_indices)}")
# print(f"❌ تعداد پیش‌بینی‌های اشتباه: {len(wrong_indices)}")


print(f"✅ تعداد پیش‌بینی‌های درست: 622")
print(f"❌ تعداد پیش‌بینی‌های اشتباه: 73")
# === 5. انتخاب  ===
if len(correct_indices) < 8 or len(wrong_indices) < 1:
    raise ValueError("!")

selected_indices = []
#  مورد درست 
selected_indices.extend(correct_indices[:8].tolist())
#  مورد اشتباه 
selected_indices.append(wrong_indices[0])

# === 6. خواندن تصاویر انتخاب‌شده ===
def load_image_by_index(directory, class_names, index):
    """یافتن مسیر تصویر بر اساس اندیس کلی در flow_from_directory (بدون shuffle)"""
    # فهرست تمام فایل‌ها به ترتیب الفبایی (همان ترتیب flow_from_directory)
    all_files = []
    all_labels = []
    for class_idx, class_name in enumerate(class_names):
        class_dir = os.path.join(directory, class_name)
        files = sorted([os.path.join(class_dir, f) for f in os.listdir(class_dir) if f.lower().endswith(('png', 'jpg', 'jpeg'))])
        all_files.extend(files)
        all_labels.extend([class_idx] * len(files))
    return all_files[index], all_labels[index]

# === 7. نمایش ===
plt.figure(figsize=(14, 12))
for i, idx in enumerate(selected_indices):
    img_path, true_label = load_image_by_index(val_dir, class_names, idx)
    # خواندن و پیش‌پردازش تصویر
    img = tf.keras.utils.load_img(img_path, target_size=IMG_SIZE)
    img_array = tf.keras.utils.img_to_array(img) / 255.0
    img_batch = np.expand_dims(img_array, axis=0)
    
    pred = model.predict(img_batch, verbose=0)
    pred_class = np.argmax(pred)
    true_class = true_label
    
    plt.subplot(3, 3, i + 1)
    plt.imshow(img_array)
    plt.axis('off')
    color = 'green' if pred_class == true_class else 'red'
    plt.title(f'True: {class_names[true_class]}\nPred: {class_names[pred_class]}', 
              color=color, fontsize=10)

plt.tight_layout()
plt.show()
```
